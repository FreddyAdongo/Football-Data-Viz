```python
df = sb.events(match_id = 3938645)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[(df['player'] == 'Jeremy Doku') & (df['type'] == 'Pass')]
```

```python
df.head()
```

```python
df = df[['location', 'pass_end_location', 'pass_outcome']]
```

```python
df.head()
```

```python
df[['x', 'y']] = df['location'].apply(pd.Series)
```

```python
df[['end_x', 'end_y']] = df['pass_end_location'].apply(pd.Series)
df['pass_outcome'] = df['pass_outcome'].fillna("Complete")
```

```python
from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb')

fig, ax = pitch.draw()

for x in df.to_dict(orient='records'):
    pitch.arrows(x['x'], x['y'],
            x['end_x'], x['end_y'], color = "#e5021c", ax=ax, width=1)

#     if x['pass_outcome'] == "Complete":
#         pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=1, 
#                    transparent=True, comet=True, ax=ax, color='g')
#     else:
#         pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=1, 
#                    transparent=True, comet=True, ax=ax, color='r')
        
ax.set_title("Jeremy Doku Passes vs Ukraine | EURO 2024",
            fontsize=12, fontweight="bold", fontfamily="monospace")
```

```python
df = sb.events(match_id = 3942226)
```

```python
df = df[df['type'] == 'Shot']
```

```python
df = df[['team', 'player', 'location', 'shot_outcome', 
         'shot_statsbomb_xg','shot_type']]
```

```python
df.head()
```

```python
df[['x', 'y']] = df['location'].apply(pd.Series)
```

```python
from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb')
fig, ax = pitch.draw()

for x in df.to_dict(orient='records'):
    if x['team'] == 'Spain':
        pitch.scatter(x=120-x['x'], y=80-x['y'], ax=ax, s=500*x['shot_statsbomb_xg'], c='#119DA4' if x['shot_outcome'] == 'Goal' else '#FFC857')
    else:
         pitch.scatter(x=x['x'], y=x['y'], ax=ax, s=500*x['shot_statsbomb_xg'], c='#119DA4' if x['shot_outcome'] == 'Goal' else '#FFC857')
            
ax.set_title('Spain vs Germany Shots | Euro 2024', fontsize=18)


import matplotlib.pyplot as plt


# # Lets also add a legend for the size of the shots in the bottom middle of the pitch
# pitch.scatter(55, 77.5, s=500*.05, ec='black', c='white', ax=ax)
# pitch.scatter(59.5, 77.5, s=500*.15, ec='black', c='white', ax=ax)
# pitch.scatter(65, 77.5, s=500*.3, ec='black', c='white', ax=ax)

# # We can add a label for the legend
# ax.text(60, 74, 'Size represents xG', ha='center', fontsize=8, fontfamily='monospace')
```
