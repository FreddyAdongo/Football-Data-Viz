```python
import requests


params = {
    'matchId': '4193903', #can switch matchId for different matches
}

response = requests.get('https://www.fotmob.com/api/matchDetails', params=params)
```

```python
data = response.json()
```

```python
shotmap = data['content']['shotmap']['shots']
```

```python
shotmap[0]
```

```python
import pandas as pd

df = pd.DataFrame(shotmap)
```
