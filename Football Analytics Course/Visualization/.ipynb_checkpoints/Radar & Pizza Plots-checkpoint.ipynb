```python
import requests
import bs4
import pandas as pd
```

```python
response = requests.get('https://fbref.com/en/comps/182/stats/NWSL-Stats')
soup = bs4.BeautifulSoup(response.content)
```

```python
comments = soup.find_all(text=lambda text:isinstance(text, bs4.Comment))
commented_out_tables = [bs4.BeautifulSoup(cmt).find_all('table') for cmt in comments]
```

```python
commented_out_tables = [tab[0] for tab in commented_out_tables if len(tab) == 1]
```

```python
df = pd.read_html(str(commented_out_tables[0]))[0]
```

```python
df.head(10)
```

```python
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]
```

```python
df.head()
```

```python
df = df.droplevel(0, axis=1)
```

```python
df.head()
```

```python
df = df[df['Pos'].str.contains('FW')]
```

```python
df.head()
```

```python
df = df[df['Min'].astype(int) > 200]
```

```python
df.head()
```

```python
df.columns
```

```python
df[['90s', 'xG', 'xAG']] = df[['90s', 'xG', 'xAG']].astype(float)
```

```python
df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']] = df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']].astype(int)
```

```python
df['goals_per_90'] = df['Gls'] / df['90s']
df['assists_per_90'] = df['Ast'] / df['90s']
df['goals_assists_per_90'] = df['G+A'] / df['90s']
df['expected_goals_per_90'] = df['xG'] / df['90s']
df['expected_assists_per_90'] = df['xAG'] / df['90s']
df['progressive_carries_per_90'] = df['PrgC'] / df['90s']
df['progressive_passes_per_90'] = df['PrgP'] / df['90s']
df['progressive_receptions_per_90'] = df['PrgR'] / df['90s']
```

```python
df['Goals'] = (df['goals_per_90'].rank(pct=True) * 100).astype(int)
df['Assists'] = (df['assists_per_90'].rank(pct=True) * 100).astype(int)
df['Goals + Assists'] = (df['goals_assists_per_90'].rank(pct=True) * 100).astype(int)
df['Expected Goals'] = (df['expected_goals_per_90'].rank(pct=True)  * 100).astype(int)
df['Expected Assists'] = (df['expected_assists_per_90'].rank(pct=True) * 100).astype(int)
df['Progressive Carries'] = (df['progressive_carries_per_90'].rank(pct=True) * 100).astype(int)
df['Progressive Passes'] = (df['progressive_passes_per_90'].rank(pct=True) * 100).astype(int)
df['Progressive Receptions'] = (df['progressive_receptions_per_90'].rank(pct=True) * 100).astype(int)
```

```python
from mplsoccer import Radar
```

```python
banda = df[df['Player'] == 'Barbra Banda']
banda.head()
```

```python
columns_to_plot = [
    'Goals', 'Assists', 'Goals + Assists',
       'Expected Goals', 'Expected Assists', 'Progressive Carries',
       'Progressive Passes', 'Progressive Receptions'
]

radar = Radar(
    params=columns_to_plot,
    min_range=[0 for _ in columns_to_plot],
    max_range=[100 for _ in columns_to_plot]
)
```

```python
fig, ax = radar.setup_axis()

rings_inner = radar.draw_circles(
    ax=ax, facecolor='white', edgecolor='black', lw=1, zorder=1
)

radar_output = radar.draw_radar(
    ax=ax, 
    values=list(banda[columns_to_plot].values.flatten()),
    kwargs_radar={'facecolor': '#77BA99'},
    kwargs_rings={'facecolor': '#EFF0D1'}
)

range_lables = radar.draw_range_labels(
    ax=ax, fontsize=15, fontproperties='monospace'
)

param_lables = radar.draw_param_labels(
    ax=ax, fontsize=15, fontproperties='monospace'
)

ax.text(x=.5, y=1.02, s='Barbra Banda', fontsize=20, ha='center', va='center',
       transform=ax.transAxes)
```

```python
chawinga = df[df['Player'] == 'Temwa Chawinga']
chawinga.head()
```

```python
fig, ax = radar.setup_axis()

rings_inner = radar.draw_circles(
    ax=ax, facecolor='white', edgecolor='black', lw=1, zorder=1
)

radar_output = radar.draw_radar_compare(
    ax=ax, 
    values=list(banda[columns_to_plot].values.flatten()),
    compare_values=list(chawinga[columns_to_plot].values.flatten()),
    kwargs_radar={'facecolor': '#E63946', 'alpha': .5},
    kwargs_compare={'facecolor': '#669BBC', 'alpha': .5}
)

range_lables = radar.draw_range_labels(
    ax=ax, fontsize=15, fontproperties='monospace'
)

param_lables = radar.draw_param_labels(
    ax=ax, fontsize=15, fontproperties='monospace'
)

ax.text(
    x=0.2, y=1.02, s='Barbra Banda', fontsize=20,
    ha='center', va='center', transform=ax.transAxes,
    fontfamily='monospace', color='#e63946'
)

ax.text(
    x=0.8, y=1.02, s='Temwa Chawinga', fontsize=20,
    ha='center', va='center', transform=ax.transAxes,
    fontfamily='monospace', color='#669bbc'
)

# Lets add some text to the lower left corner
ax.text(
    x=0, y=0.05, s='Metrics show per 90 percentile stats\n\nComparing against all forwards\nwith at least 200 minutes played\nData: FBREF', fontsize=10,
    ha='left', va='center', transform=ax.transAxes, fontfamily='monospace'
)
```

```python
from mplsoccer import PyPizza
```

```python
pizza = PyPizza(
    params=columns_to_plot,
    background_color='#FEFEFF',
    straight_line_color='#000807',
    straight_line_lw=1,
    last_circle_lw=1,
    other_circle_lw=0,
    other_circle_color='#000807',
    other_circle_ls='--'
)
```

```python
fig, ax = pizza.make_pizza(
    figsize=(8, 8),
    values=banda[columns_to_plot].values.flatten().tolist(),
    kwargs_slices=dict(
        facecolor='#414073', edgecolor='#000807', linewidth=1
    ),
    kwargs_params=dict(
        color='#000807', fontsize=11, fontproperties="monospace"
    ),
    kwargs_values=dict(
        color='#EFF2F1', fontsize=11, fontproperties="monospace", 
        bbox = {
            "edgecolor": "#000807",
            "facecolor": "#414073",
            "boxstyle": "round,pad=0.2",
            "lw": 1
        }
    )
)

# add title
fig.text(
    0.515, 0.98, "Barbra Banda", size=18,
    ha="center", fontfamily="monospace", color="#000000"
)

# add subtitle
fig.text(
    0.515, 0.952,
    "Per 90 Percentile Rank vs NWSL Forwards | 2024",
    size=11,
    ha="center", fontfamily="monospace", color="#000000"
)

CREDIT_1 = "Data: FBREF"
CREDIT_2 = "@freddy_on_canvas"

fig.text(
    0.075, 0.005, f"{CREDIT_1}\n{CREDIT_2}", size=9,
    fontfamily="monospace", color="#000000",
    ha="left"
)
```
