```python
df = df.rename(columns={'Unnamed: 1_level_0 Player' : 'Player'}) #rename column
```

```python
df = df.rename(columns=lambda x: x.split(' ')[-1] if 'Unnamed' in x else x)
```

```python
#filter a dataframe
prem = df[df['Comp'] == 'eng Premier League']
prem.head()
```

```python
#cannot compare string to ints
#prem = df[(df['Comp'] == 'eng Premier League') & (df['Playing Time MP'] > 20)]
#print(prem.head())
```

```python
#drop repeated columns
df = df.drop(df[df['Player'] == 'Player'].index)
```

```python
#now we typecast the 'Playing Time MP' column to an int
df['Playing Time MP'] = df['Playing Time MP'].astype(int)
```

```python
df[['Performance Gls', 'Performance.1 Ast']] = df[['Performance Gls', 'Performance.1 Ast']].astype(int)
```

```python
df.dtypes
```

```python
prem = df[(df['Comp'] == 'eng Premier League') & (df['Playing Time MP'] > 20)]
print(prem.head())
```

```python
#sort column values
df.sort_values(by=['Age', 'Squad'], ascending=False)
```

```python
prem['Minutes per Goal'] = prem['Playing Time.2 Min'].astype(int) / prem['Performance Gls'].astype(int)
```

```python
#calculate Minutes per Goal
prem[['Player', 'Minutes per Goal']].sort_values('Minutes per Goal').head(10)
```

```python
#group by
df['Age'] = df['Age'].astype(int)

df.groupby('Comp')['Age'].mean() #calculating average age in the following leagues
```

```python
#.where() method
df['Age Group'] = np.where(
    df['Age'] < 25,
    'Under 25', 
    'Not Under 25'
)

df[['Player', 'Age', 'Age Group']].head(15) #shows players Under 25 and Not
```

```python
import matplotlib.pyplot as plt

df[['Playing Time.2 Min', 'Performance Gls']] = df[['Playing Time.2 Min', 'Performance Gls']].astype(int)

x = df['Playing Time.2 Min']
y = df['Performance Gls']

m, b = np.polyfit(x, y, 1) #line of best fit

plt.scatter(x, y)
plt.plot(x, m*x + b, color='red') #calculate slope
```
